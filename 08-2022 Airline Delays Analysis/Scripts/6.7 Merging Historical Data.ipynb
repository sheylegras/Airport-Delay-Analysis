## 1. Importing Libraries and Data

```python
#Import libraries
import pandas as pd
import numpy as np
import os
```

```python
#define path
path = r'C:\Users\legra\Desktop\08-2022 Airline Delays Analysis'
```

```python
#import data
df = pd.read_csv(os.path.join(path, 'Data', 'Prepared Data', 'wrangled_v3.csv'))
```

## 2. Data Cleaning

```python
df.head()
```

```python
df =df.drop(columns='Unnamed: 0')
```

```python
df.head()
```

```python
df_year = df.groupby('year').sum()
```

```python
df_year['percent_carrier_delay_minutes'] = df_year['carrier_delay_minutes'] / df_year['arrival_delay_minutes']*100
df_year['percent_weather_delay_minutes'] =df_year['weather_delay_minutes'] / df_year['arrival_delay_minutes']*100
df_year['percent_nas_delay_minutes'] = df_year['nas_delay_minutes'] / df_year['arrival_delay_minutes']*100
df_year['percent_security_delay_minutes'] = df_year['security_delay_minutes'] / df_year['arrival_delay_minutes']*100
df_year['percent_late_aircraft_delay_minutes'] = df_year['late_aircraft_delay_minutes'] / df_year['arrival_delay_minutes']*100
```

```python
df_year.head()
```

```python
df_year.columns
```

```python
#create subset
df_sub = df_year[['percent_carrier_delay_minutes', 'percent_weather_delay_minutes', 'percent_nas_delay_minutes', 'percent_security_delay_minutes', 'percent_late_aircraft_delay_minutes']]
```

```python
df_sub.head()
```

```python
#reset index
df_sub = df_sub.reset_index()
```

```python
df_sub.head()
```

```python
df_sub.shape
```

## 3. Merging Data

```python
#import historical data to merge
df_historical = pd.read_csv(os.path.join(path, 'Data', 'Original Data', 'delay_cause_by_year.csv')) 
```

```python
df_historical.head(20)
```

```python
#dropping observations
df2 = df_historical.drop([0,17])
```

```python
df2.columns
```

```python
#rename columns
df_sub2 = df2.rename(columns ={'Unnamed: 0' : 'year', 'Air Carrier Delay' : 'percent_carrier_delay_minutes', 'Aircraft Arriving Late' : 'percent_late_aircraft_delay_minutes', 'National Aviation System Delay' : 'percent_nas_delay_minutes',
                    'Security Delay' : 'percent_security_delay_minutes', 'Extreme Weather' : 'percent_weather_delay_minutes'})
```

```python
df_sub2.shape
```

```python
#concat data
frames = [df_sub, df_sub2]
df_concat = pd.concat(frames)
```

```python
df_concat
```

```python
#xport df
df_concat.to_csv(os.path.join(path, 'Data', 'Prepared Data', 'historical.csv'))
```
