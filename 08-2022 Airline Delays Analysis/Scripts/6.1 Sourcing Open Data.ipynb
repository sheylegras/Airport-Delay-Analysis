## 04. Checking for Missing Data

```python
#checking for missing values
df.isnull().sum()
```

```python
#create subset to view missing values
df_nan = df[df['arr_flights'].isnull() == True]
```

```python
#checking output
df_nan
```

Additional missing data from the arr_del15 column represent flights that were cancelled.

```python
#create df without missing values
df_clean = df[df['arr_flights'].isnull() == False]
```

```python
#checking for accuracy
df_clean.isnull().sum()
```

```python
df_clean.shape
```

## 05. Checking for Duplicates

```python
#create df for dups
df_dups = df_clean[df_clean.duplicated()]
```

```python
df_dups
```

No duplicates found

## 06. Checking for Mixed Datatypes

```python
for col in df_clean.columns.tolist():
  weird = (df_clean[[col]].applymap(type) != df_clean[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_clean[weird]) > 0:
    print (col)
```

No mixed data types found

## 07. Data Cleaning

```python
#changing column names for clarity
df_clean.rename(columns = {'arr_flights' : 'arriving_flight_count'}, inplace = True)
df_clean.rename(columns = {'arr_del15' : 'arrival_delayed_15_minutes_more'}, inplace = True)
df_clean.rename(columns = {'carrier_ct' : 'carrier_delay_count'}, inplace = True)
df_clean.rename(columns = {'weather_ct' : 'weather_delay_count'}, inplace = True)
df_clean.rename(columns = {'nas_ct' : 'nas_delay_count'}, inplace = True)
df_clean.rename(columns = {'security_ct' : 'security_delay_count'}, inplace = True)
df_clean.rename(columns = {'late_aircraft_ct' : 'late_aircraft_count'}, inplace = True)
df_clean.rename(columns = {'arr_cancelled' : 'arrival_cancelled_count'}, inplace = True)
df_clean.rename(columns = {'arr_diverted' : 'arrival_diverted_count'}, inplace = True)
df_clean.rename(columns = {'arr_delay' : 'arrival_delay_minutes'}, inplace = True)
df_clean.rename(columns = {'carrier_delay' : 'carrier_delay_minutes'}, inplace = True)
df_clean.rename(columns = {'weather_delay' : 'weather_delay_minutes'}, inplace = True)
df_clean.rename(columns = {'nas_delay' : 'nas_delay_minutes'}, inplace = True)
df_clean.rename(columns = {'security_delay' : 'security_delay_minutes'}, inplace = True)
df_clean.rename(columns = {'late_aircraft_delay' : 'late_aircraft_delay_minutes'}, inplace = True)
```
